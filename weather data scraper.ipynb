```python
#import library and website

import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.36 Safari/537.36'}
url = 'https://www.accuweather.com/en/us/san-francisco/94103/hourly-weather-forecast/347629?day=3'
r = requests.get(url, headers = headers)
html_doc = r.text
```

```python
# scrape weather type and cloud coverage
soup = BeautifulSoup(html_doc, 'html.parser')
weather_type = soup.findAll('span', {"class": ["phrase"]})
cloud_coverage = soup.findAll('div', {"class": ["panel"]})
date_time = soup.findAll('div', {"class": ["date"]})
```

```python
#get single cloud coverage
panel = []
for i in range(1, len(cloud_coverage), 2):
    panel.append(cloud_coverage[i])

cloud_cover = []
for i in range(0, len(panel)):
    panel_elem = BeautifulSoup(str(panel[i]), "html.parser")
    temp_cloud_cover = panel_elem.findAll('p')
    cloud_cover.append(str(temp_cloud_cover[0])[23:25])
```

```python
#get weather type

weather = []
for i in range(0, len(weather_type)):
    weather_temp = BeautifulSoup(str(weather_type[i]), "html.parser")
    weather.append(str(weather_temp)[25:-10])
```

```python
#get time and date
from datetime import datetime
time = []
for i in range(0, len(date_time)):
    date_time_temp = BeautifulSoup(str(date_time[i]), "html.parser")
    time_temp = date_time_temp.findAll('p')[0].text.replace('\n', '').replace('\t', '')
    date_temp = date_time_temp.findAll('p')[1].text.replace('\n', '').replace('\t', '')
    if date_temp[1] == "/":
        date_temp = "2020/0" + date_temp
    if time_temp[-2:] == "PM":
        time_temp = str(int(time_temp[:-3]) + 12)
        if time_temp == "24":
            time_temp = "12"
        time_temp += ":30:00"
    if time_temp[-2:] == "AM":
        time_temp = time_temp[:-3]
        if time_temp == "12":
            time_temp = "00"        
        time_temp += ":30:00"
    final = time_temp + ' ' + date_temp
    from datetime import datetime
    myDate = datetime.strptime(final, "%H:%M:%S %Y/%m/%d")
    time.append(myDate)
```

```python
dict_all = {'Time': time, 'Weather Type': weather, 'Cloud Cover (%)': cloud_cover}
df = pd.DataFrame(dict_all)
```
